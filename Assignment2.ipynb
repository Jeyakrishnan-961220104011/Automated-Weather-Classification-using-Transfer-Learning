```python

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/House Price India.csv')
df.head()
     
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# remove unwanted column

df.drop(['id','Date'],axis =1,inplace =True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.Lattitude.value_counts()
```

```python
x= df.drop(['Lattitude'],axis =1)
y = df.Lattitude
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
     
```

```python
y = le.fit_transform(y)
```

```python
y
```

```python
# split the data (Training & Testing)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
# creating ANN Structure
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

classification = Sequential()
classification.add(Dense(30,activation = 'relu'))
classification.add(Dense(128,activation = 'relu'))
classification.add(Dense(64,activation = 'relu'))
classification.add(Dense(32,activation = 'relu'))
classification.add(Dense(1,activation = 'sigmoid'))
```

```python
# compiling the ANN model

classification.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
# Training the model

classification.fit(xtrain,ytrain,batch_size = 10,epochs = 300)
```

```python
# Testing the model

y_pred = classification.predict(xtest)
```

```python
pd.DataFrame({'Actual value':ytest.flatten(),'predicted value':y_pred.flatten().astype('int')}).head(10)
```
